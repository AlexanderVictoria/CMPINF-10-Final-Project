## Introduction
One aspect that makes a neighborhood safe is the risk of fires from other buildings spreading throughout the community. A good way to understand if a certain community is at risk of this is to look at all the fire reports for every community in Pittsburgh. Using the fire-reports.csv that was obtained from the WPRDC's online database, I will find the safest community at least risk of fire damage based one number of reports.

### Import Libraies

```python
import pandas as pd
```

### Load Data

```python
fire = pd.read_csv("fire-reports.csv")
```

### Sort and Process Data

```python
fire_sorted = fire.groupby(fire['neighborhood']).size()
fire_sorted.sort_values(ascending=True, inplace=True) 
fire_sorted.plot(kind='bar', figsize=(15, 6))
```

### Top 20 Neighborhoods

```python
top_fire = fire_sorted.head(20)
top_fire
```

```python
fire_plot = top_fire.plot(kind='bar')
```

```python
fire_worst = fire_sorted.tail(20)
fire_worst
```

```python
fire_worst_plot = fire_worst.plot(kind='bar')
```

## Conclusion

* The best neighborhood based on number of fire reports is **Mount Oliver Borough** with the **6** fire reports from 2014 to 2016
* The worst neighborhood based on number of fire reports is **Carrick** with **307** fire reports from 2014 to 2016
